```python
import pandas as pd
import joblib
from sklearn.preprocessing import LabelEncoder
import os

# [수정 포인트] 노트북이 이미 데이터 폴더 안에 있으므로, 경로는 현재 위치('./')로 설정합니다.
data_path = 'data/'

# 디버깅: 현재 위치와 파일 목록을 출력해 봅니다.
print(f"현재 작업 경로: {os.getcwd()}")
print(f"경로 내 파일 목록: {os.listdir(data_path)}")

# 1. 데이터 및 모델 로드
try:
    # 경로 앞에 data_path를 붙여서 파일을 읽습니다. (./test.csv)
    df_test = pd.read_csv(data_path + 'test.csv')
    model = joblib.load(data_path + 'model.pkl')
    print("✅ 데이터와 모델을 성공적으로 불러왔습니다.")
except FileNotFoundError as e:
    print(f"❌ 파일을 찾을 수 없습니다. 경로를 다시 확인해주세요: {e}")
    raise

# 2. 전처리 (연구자 1의 train.py와 동일한 로직 적용)
# 범주형 데이터 변환 (Extracurricular Activities: Yes/No -> 1/0)
le = LabelEncoder()
if 'Extracurricular Activities' in df_test.columns:
    df_test['Extracurricular Activities'] = le.fit_transform(df_test['Extracurricular Activities'])

# 학습 모델이 기대하는 Feature 순서대로 데이터 선택
features = ['Hours Studied', 'Previous Scores', 'Extracurricular Activities', 'Sleep Hours', 'Sample Question Papers Practiced']
X_test = df_test[features]

# 3. 추론 수행
predictions = model.predict(X_test)

# 4. 결과 저장
# 예측 결과를 DataFrame에 추가
df_test['Predicted Performance'] = predictions

# result.csv 파일로 저장
save_path = data_path + 'result.csv'
df_test.to_csv(save_path, index=False)

print(f"🎉 추론 완료! 결과가 '{save_path}' 파일로 저장되었습니다.")
print("▼ 예측 결과 미리보기:")
print(df_test[['Predicted Performance']].head())
```
